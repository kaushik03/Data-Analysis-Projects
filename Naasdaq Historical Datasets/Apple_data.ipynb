```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
from pandas_profiling import ProfileReport
```

```python
apl_df = pd.read_csv("./Data/AAPL_naasdaq.csv")
apl_df.head()
```

```python
#apl_df['Date'].isna().sum()
apl_df.rename(columns={' Close/Last':"Close"},inplace = True)
apl_df.head()
```

```python
apl_df.set_index('Date')
```

```python
profile = ProfileReport(apl_df, title='Apple Stocks reports', explorative=True)
```

```python
apl_df.info()
```

```python
close = []
y = apl_df['Close']
for c in y:
    c=c.replace('$','')
    close.append(float(c))
```

```python
x = apl_df['Date']
dates = []
import datetime
import time

for i in x:
    val = time.mktime(datetime.datetime.strptime(i,'%m/%d/%Y').timetuple())
    dates.append(val)
```

```python
ax = sns.lineplot(dates,close)
ax._legend.set_title("Closing Price in Dollars")
```
