## Clearance Data

This data is from Jan 2019 - Dec 2019 regarding the clearance of consignments of various shipments.

```python
%pylab inline
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("Clearance Data.csv")
data.head()
```

```python
data['Clearance Date'] = pd.to_datetime(data['Clearance Date'], format= '%d-%m-%Y')
```

```python
data['Invoice Date'] = pd.to_datetime(data['Invoice Date'], format= '%d-%m-%Y',errors = 'ignore')
data.dtypes
```
