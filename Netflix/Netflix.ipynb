# Netflix Data 

This data was posted on [kaggle](kaggle.com). and the link is [here](https://www.kaggle.com/shivamb/netflix-shows)

This data contains information on TV Shows/Movies released and origin.

```python
%pylab inline
import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("./Data/netflix_titles.csv")
df.head()
```

## Cleaning Data as per requirement

```python
df.info()
```

```python
df.set_index('show_id',inplace = True)
```

```python
df.isnull().sum()
```

```python
#filling All Nulls values with a value
df['director'].fillna("NA",inplace = True)
df['cast'].fillna("NA",inplace= True) 
df['rating'].fillna("NA",inplace = True)
df['country'].fillna('NA', inplace = True)
df['date_added'].fillna(datetime.date.today(),inplace = True)
df['date_added'] = pd.to_datetime(df['date_added'], infer_datetime_format=True)
df['release_year'] = pd.to_datetime(df['release_year'], format = '%Y')
df['release_year'] = df['release_year'].dt.to_period('Y')
df.head(10)
```

```python
df.rename(columns = {"type":"Category","title":"Title","director":"Director","cast": "Cast","country":"Country","date_added":"Date Added","release_year":"Release Year","rating":"Rating","duration":"Duration","listed_in":"Genres"},inplace = True)
```

```python
df.shape
```

```python
df.drop(['description'],axis = 1,inplace = True)
```

```python
df.info()
```

```python
df = df.sort_values('Release Year',ascending = True)
df.head()
```
