```python
%pylab inline
import pandas
import seaborn
import matplotlib.pyplot as plt
```

```python
data.plot(kind='bar', figsize=(20,20), x='Country/Region', y ='Deaths')
plt.xlabel("Country by their Name", labelpad=14)
plt.ylabel("Confirmed Cases worldwide", labelpad=14)
plt.show()
```
