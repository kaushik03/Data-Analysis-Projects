```python
import pandas as pd
import pandas_bokeh
import datetime as dt
pd.set_option("plotting.backend","pandas_bokeh")
pd.plotting.output_notebook()
```

```python
df = pd.read_csv("./Data/India/covid_vaccine_statewise.csv")
df.head()
```

```python
df.info()
```

```python
df["Updated On"] = pd.to_datetime(df["Updated On"])
df["Month"] = pd.to_datetime(df["Updated On"]).dt.strftime("%m")
df.head()
```

```python
df.Month.value_counts()
```

```python
Vaccine_injan = df[df["Month"]=="01"]
Vaccine_injan.info()
```

```python
Vaccination_in_Mah = df[df["State"] == "Maharashtra"].groupby("Month",as_index = False).mean()
Vaccination_in_Mah
```

```python
y_axis = ["Total Individuals Vaccinated","Male(Individuals Vaccinated)","Female(Individuals Vaccinated)","18-45 years (Age)",
         "45-60 years (Age)","60+ years (Age)"]
plot_Mah = Vaccination_in_Mah.plot.line(x = "Month",
                                       y = y_axis,
                                       plot_data_points = True,
                                       plot_data_points_size = 5,
                                       marker = "circle")
```
