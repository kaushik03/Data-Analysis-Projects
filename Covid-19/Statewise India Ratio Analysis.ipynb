```python
import pandas as pd
import pandas_bokeh
pd.set_option("plotting.backend","pandas_bokeh")
pd.plotting.output_notebook()
```

```python
df = pd.read_csv("./Data/India/Latest Covid-19 India Status.csv")
df.head()
```

```python
State_list = ["Maharashtra","Gujarat","Uttar Pradesh","Delhi","Karnataka","Goa","Bihar"]
mod_df = df[df["State/UTs"].isin(State_list)]
mod_df
```

```python
Ratio_plot = mod_df.plot.bar(x = "State/UTs",
                      y = ["Active Ratio","Discharge Ratio","Death Ratio"],
                       xlabel = "States",
                       ylabel = "Ratios",
                       title = "Latest Status of The Following State",
                      show_figure = False)

Cases_plot = mod_df.plot.bar(x = "State/UTs",
                      y = ["Active","Discharged","Deaths"],
                       xlabel = "States",
                       ylabel = "Ratios",
                       title = "Latest Status of The Following State",
                      show_figure = False)
Total_cases_plot = mod_df.plot.line(x = "State/UTs",
                      y = "Total Cases",
                       xlabel = "States",
                       ylabel = "Total Cases",
                       title = "Total Cases of The Following State",
                        plot_data_points = True,
                        plot_data_points_size = 5,
                        marker = "circle",
                        zooming = False,
                      show_figure = False)
```

```python
Total_cases_plot.plot_width = 900
dashboard = pandas_bokeh.column(Total_cases_plot,
                pandas_bokeh.row(Cases_plot, Ratio_plot))
pandas_bokeh.show(dashboard)
```
