## Importing Data

### Changing value of Gender in "Sex Column" for better interpretation and Understanding

- 1 : M -> Male
- 2 : F -> Female

```python
df["Sex"].replace(to_replace=[1,2], value= ["M","F"], inplace=True)
```

```python
df.head()
```

```python
df.drop(["Admin1", "SubDiv"], axis = 1)
```

## Combining data according to WHO AGE Group Standard
### Age Group

-  Deaths2 to Deaths7 = Kids
-  Deaths8 to Deaths9 = Teenage
-  Deaths10 to Deaths11 = Young_Adults
-  Deaths12 to Deaths15 = Middle_aged
-  Deaths16 to Deaths25 = Old_aged
-  Deaths26 = Unspecified
-  IM_deaths1-4 = Infants (0 to 1yr)

```python
df.fillna(0, inplace = True)
```

```python
df["Infants"] = df['IM_Deaths1']+df['IM_Deaths2']+df['IM_Deaths3']+df['IM_Deaths4']
df["Kids"] = df['Deaths2']+df['Deaths3']+df['Deaths4']+df['Deaths5']+df['Deaths6']+df['Deaths7']
df["Teenage"] = df['Deaths8']+df['Deaths9']
df["Young_Adults"] = df['Deaths10']+df['Deaths11']
df["Middle_aged"] = df['Deaths12']+df['Deaths13']+df['Deaths14']+df['Deaths15']
df["Old_aged"] = df['Deaths16']+df['Deaths17']+df['Deaths18']+df['Deaths19']+df['Deaths20']+df['Deaths21']+df['Deaths22']+df['Deaths23']+df['Deaths24']+df['Deaths25']
df.rename(columns={"Deaths26": "Unspecified", "Deaths1": "All_Deaths"})
df.head()
```

```python
df.drop(["Deaths2","Deaths3","Deaths4","Deaths5","Deaths6","Deaths7","Deaths8","Deaths9","Deaths10","Deaths11","Deaths12"
        ,"Deaths13","Deaths14","Deaths15","Deaths16","Deaths17","Deaths18","Deaths19","Deaths20","Deaths21","Deaths22","Deaths23","Deaths24","Deaths25","Deaths26"],axis = 1, inplace =True)
```

```python
df.drop(["IM_Deaths1","IM_Deaths2","IM_Deaths3","IM_Deaths4"], axis = 1, inplace = True)
df.head()
```

```python
df.to_csv("WHO_mortality_cleaned.csv")
```
